# Simple Thresholding

Applies a fixed-level [threshold](https://docs.opencv.org/2.4/modules/imgproc/doc/miscellaneous_transformations.html?highlight=cv2.threshold#threshold) to each array element. It can be done as follows,

```cv2.threshold(Source, threshold value, maximum value, type[, Output]) → retval, Output```

Parameters:	
* **Source**: input array (single-channel, 8-bit or 32-bit floating point).
* **Output**: output array of the same size and type as Source.
* **threshold value**: threshold value.
* **maximum value** – maximum value to use with the THRESH_BINARY and THRESH_BINARY_INV thresholding types.
* type:
    * THRESH_BINARY
    * THRESH_BINARY_INV
    * THRESH_TRUNC
    * THRESH_TOZERO
    * THRESH_TOZERO_INV

## THRESH_BINARY
$$\text{Output }(x,y) = \begin{cases}\text{maximum value}&\mbox{if source(x,y) > threshold value},\\0,&\mbox{otherwise}\end{cases}$$

````python
_ , Out = cv2.threshold(Img, 130, 255,cv2.THRESH_BINARY)

fig, ax = plt.subplots(1, 2 , figsize = (17, 6))
ax = ax.ravel()
font = FontProperties()
font.set_weight('bold')
_ = fig.suptitle('THRESH_BINARY', fontproperties=font, fontsize = 18)

_ = ax[0].imshow(Img)
_ = ax[0].set_title('Original Image', fontproperties=font, fontsize = 12)
_ = ax[1].imshow(Out)
_ = ax[1].set_title('Threshold value = 130,  Maximum value = 255', fontproperties=font, fontsize = 12)
for i in range(2):
    _ = ax[i].set_aspect('auto')
    _ = ax[i].axis('off')
del i
````

## THRESH_BINARY_INV
$$\text{Output} (x,y) = \begin{cases}0&\mbox{if source(x,y) > threshold value},\\\text{maximum value},&\mbox{otherwise}\end{cases}$$

````python
_ , Out = cv2.threshold(Img, 130, 255,cv2.THRESH_BINARY_INV)

fig, ax = plt.subplots(1, 2 , figsize = (17, 6))
ax = ax.ravel()
font = FontProperties()
font.set_weight('bold')
_ = fig.suptitle('THRESH_BINARY_INV', fontproperties=font, fontsize = 18)

_ = ax[0].imshow(Img)
_ = ax[0].set_title('Original Image', fontproperties=font, fontsize = 12)
_ = ax[1].imshow(Out)
_ = ax[1].set_title('Threshold value = 130,  Maximum value = 255', fontproperties=font, fontsize = 12)
for i in range(2):
    _ = ax[i].set_aspect('auto')
    _ = ax[i].axis('off')
del i
````

## THRESH_TRUNC
$$\text{Output} (x,y) = \begin{cases}\text{threshhold}&\mbox{if source(x,y) > tthreshold value},\\\text{source(x,y)},&\mbox{otherwise}\end{cases}$$

````python
_ , Out = cv2.threshold(Img, thresh = 130, maxval = 255, type = cv2.THRESH_TRUNC)

fig, ax = plt.subplots(1, 2 , figsize = (17, 6))
ax = ax.ravel()
font = FontProperties()
font.set_weight('bold')
_ = fig.suptitle('THRESH_TRUNC', fontproperties=font, fontsize = 18)

_ = ax[0].imshow(Img)
_ = ax[0].set_title('Original Image', fontproperties=font, fontsize = 12)
_ = ax[1].imshow(Out)
_ = ax[1].set_title('Threshold value = 130', fontproperties=font, fontsize = 12)
for i in range(2):
    _ = ax[i].set_aspect('auto')
    _ = ax[i].axis('off')
del i
````

## THRESH_TOZERO

# AdaptiveThresholding

Applies an adaptive threshold [threshold](https://docs.opencv.org/2.4/modules/imgproc/doc/miscellaneous_transformations.html?highlight=adaptivethreshold) to an array.. It can be done as follows,

```cv2.adaptiveThreshold(Source, Maximum Value, Adaptive Method, thresholdType, blockSize, C[, Output]) → Output```

Parameters:	
* **Source**: Source 8-bit single-channel image.
* **Output**: Destination image of the same size and the same type as Source.
* **Maximum Value** – Non-zero value assigned to the pixels for which the condition is satisfied. See the details below.
* **Adaptive Method**: Adaptive thresholding algorithm to use, ADAPTIVE_THRESH_MEAN_C or ADAPTIVE_THRESH_GAUSSIAN_C . See the details below.
* thresholdType:
    * THRESH_BINARY
    * THRESH_BINARY_INV
* **blockSize**: Size of a pixel neighborhood that is used to calculate a threshold value for the pixel: 3, 5, 7, and so on.
* **C**: Constant subtracted from the mean or weighted mean (see the details below). Normally, it is positive but may be zero or negative as well.

## THRESH_BINARY
$$\text{Output }(x,y) = \begin{cases}\text{maximum value}&\mbox{if source(x,y)} > T(x,y),\\0,&\mbox{otherwise}\end{cases}$$

````python
Out = cv2.adaptiveThreshold(cv2.cvtColor(Img,cv2.COLOR_BGR2GRAY), 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,11,2)

fig, ax = plt.subplots(1, 2 , figsize = (17, 6))
ax = ax.ravel()
font = FontProperties()
font.set_weight('bold')
_ = fig.suptitle('THRESH_BINARY', fontproperties=font, fontsize = 18)

_ = ax[0].imshow(Img)
_ = ax[0].set_title('Original Image', fontproperties=font, fontsize = 12)
_ = ax[1].imshow(Out,'gray')
_ = ax[1].set_title('Maximum value = 255', fontproperties=font, fontsize = 12)
for i in range(2):
    _ = ax[i].set_aspect('auto')
    _ = ax[i].axis('off')
del i
````

## THRESH_BINARY_INV

$$\text{Output }(x,y) = \begin{cases}0&\mbox{if source(x,y)} > T(x,y),\\\text{maximum value},&\mbox{otherwise}\end{cases}$$

````python
Out = cv2.adaptiveThreshold(cv2.cvtColor(Img,cv2.COLOR_BGR2GRAY), 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV,11,2)

fig, ax = plt.subplots(1, 2 , figsize = (17, 6))
ax = ax.ravel()
font = FontProperties()
font.set_weight('bold')
_ = fig.suptitle('THRESH_BINARY_INV', fontproperties=font, fontsize = 18)

_ = ax[0].imshow(Img)
_ = ax[0].set_title('Original Image', fontproperties=font, fontsize = 12)
_ = ax[1].imshow(Out,'gray')
_ = ax[1].set_title('Maximum value = 255', fontproperties=font, fontsize = 12)
for i in range(2):
    _ = ax[i].set_aspect('auto')
    _ = ax[i].axis('off')
del i
````
